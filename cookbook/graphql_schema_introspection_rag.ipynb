```python
import os
import json
from langchain_core.embeddings import Embeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import JSONLoader
from langchain_core.vectorstores import InMemoryVectorStore
from pathlib import Path
from pprint import pprint
from langchain_openai import OpenAIEmbeddings
```

```python
#load schema
with open("../queries.json", "r") as file:
    content = json.load(file)
    schema = content['data']['__schema']['queryType']['fields']
```

```python
# Load schema to JSONLoader
loader = JSONLoader(
    file_path='../queries.json',
    jq_schema='.data.__schema.queryType.fields[].name',
    text_content=False)

data = loader.load()
```

```python
# load Documents to InMemoryVectorStore
vector_store = InMemoryVectorStore(OpenAIEmbeddings())
```

```python
vector_store.add_documents(documents=data)
```
